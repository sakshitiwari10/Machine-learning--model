```python
### Creating arrays
```

```python
a = np.array([1,2,3,4])
print(a)
```

```python
b = np.array([[1,2,3,4],[5,6,7,8]]) # 2-d array
print(b)
```

```python
a = np.array([0,0,0,0])
print(a)
```

```python
zero = np.zeros((2,3))
print(zero)
```

```python
print(np.ones((2,3)))
```

```python
print(np.full((2,3),7))
```

```python
print(np.arange(0,10,2))
```

```python
print(np.eye(3))
```

```python
rng = np.random.default_rng(seed=42)
print(rng.random((2,3)))

```

```python
print(rng.integers(1,10,size=(3,3))) # uniform
```

```python
print(rng.normal(loc=0,scale=1,size=5))
```

```python
### array attributes and data types
```

```python
M = np.array ([[1,2,3],[4,5,6]])
print(M)
print(M.ndim)   # Number of dimensions
print(M.shape)  # Shape of the array
print(M.size)
print(M.dtype)
```

```python
ints = np.array([1,2,3], dtype=np.int32)
```

```python
ints.dtype
```

```python
floats = np.array([1,2,3],dtype = np.float64)
floats.dtype
```
