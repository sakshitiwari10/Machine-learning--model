```python
# if else if
```

```python
marks = int(input("enter your marks here :"))
if marks >= 90:
    print("A")
elif marks >=80:
    print("B")
elif marks >=70:
    print("C")
else:
    print("Fail")
```

### List Comprehnsion

```python
 numbers = [1,2,3,4,5]
square = []

for x in numbers:
    square.append(x**2)
print(square)
```

```python
square = [x**2 for x in range(5)]
print(square)
```

```python
even = [x for x in range(20) if x % 2 ==0]
print(even)
```

```python
word = ["python", "machine learning", "dataset"]
upper = [w.upper() for w in word]
print(upper) 
```

### Dictionary


```python
student ={"name":"Sakshi","age":22,"marks":92}
print(student["marks"])
```

```python
# add , update, delete
```

```python
student = {"name":"Sakshi","age":22}
student["city"] = "Surat"
print(student)
```

```python
student = ['age']
del student   # deletes the variable
```

```python
marks = {"maths":90,"phy":85,"chem":82}
for subject,score in marks.items():
    print(subject,score)
for score in marks.values():
    print(score)
```
