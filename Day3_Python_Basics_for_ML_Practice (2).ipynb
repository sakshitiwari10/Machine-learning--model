### 19. Mini Challenge

**Question:** Write a program that takes a list of marks `[45, 78, 89, 56, 92, 67]` and prints: the average, the highest, the lowest, and a list of marks that are above the average (use list comprehension).

```python

marks = [45, 78, 89, 56, 92, 67]

average = sum(marks) / len(marks)

highest = max(marks)
lowest = min(marks)
above_average = [m for m in marks if m > average]

# Print results
print("Average:", average)
print("Highest:", highest)
print("Lowest:", lowest)
print("Marks above average:", above_average)

```
